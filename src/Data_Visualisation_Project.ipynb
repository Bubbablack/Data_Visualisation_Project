```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns

training = pd.read_csv('../data/training.csv')
training.head()
```

```python
#this is for males
noob = len(training.loc[(training['mobile_money_classification'] == 0)& (training['Q2'] == 1)])
nomob = len(training.loc[(training['mobile_money_classification'] == 1) & (training['Q2'] == 1)])
mob =len( training.loc[(training['mobile_money_classification']==2) & (training['Q2'] == 1)])
both = len(training.loc[(training['mobile_money_classification']==3 )&(training['Q2'] == 1)])
#this is for the females
noob2 = len(training.loc[(training['mobile_money_classification'] == 0)& (training['Q2'] == 2)])
nomob2 = len(training.loc[(training['mobile_money_classification'] == 1) & (training['Q2'] == 2)])
mob2 =len( training.loc[(training['mobile_money_classification']==2) & (training['Q2'] == 2)])
both2 = len(training.loc[(training['mobile_money_classification']==3 )&(training['Q2'] == 2)])


somting= ['non-mobile','mobile','both']
xtings = [nomob,mob,both]
females = [nomob2,mob2,both2]
fig, ax = plt.subplots()

ax.bar(somting, xtings,0.35, label='Men')
ax.bar(somting, females, 0.35, bottom=xtings,label='Women')
ax.set_ylabel('number of participents')
ax.set_title('Relationship between gender and banking services')
ax.legend()
plt.show()
```

```python
nomobile = training.loc[training['mobile_money_classification'] == 1]
mobile = training.loc[training['mobile_money_classification'] == 2]
have_both = training.loc[training['mobile_money_classification'] == 3]


nonmobile_status = nomobile.groupby('Q3').Q3.count()
mobile_status = mobile.groupby('Q3').Q3.count()
both_status = have_both.groupby('Q3').Q3.count()
N = 4
width = 0.25
status= np.arange(N)

plt.bar(status, nonmobile_status,width , label='nonmobile')
plt.bar(status+width, mobile_status, width , label='mobile')
plt.bar(status + 2 * width, both_status,width, label='both')

plt.ylabel('number of people')
plt.xlabel('marital status')
plt.title('non-mobile service users grouped by marital status')

plt.xticks(status + width , ('maried', 'devorced', 'widowed', 'single'))
plt.legend(loc='best')
plt.show()
```

```python
nonmobile_status = nomobile.groupby('Q5').Q5.count()
mobile_status = mobile.groupby('Q5').Q5.count()
both_status = have_both.groupby('Q5').Q5.count()
N = 6
width = 0.25
status= np.arange(N)

plt.bar(status, nonmobile_status, width , label='nonmobile')
plt.bar(status+width, mobile_status, width , label='mobile')
plt.bar(status + 2 * width, both_status,width, label='both')

plt.ylabel('number of people')
plt.xlabel('land ownership status')
plt.title('land ownership status')

plt.xticks(status + width , ('personally','someon', 'member', ' renting','not','dont know' ))
plt.legend(loc='best')
plt.show()
```
